```python
from ultralytics import YOLO
import torch
# import torch_directml
import os

def train_model():

    model = YOLO("yolo26s.pt")


    device = "cuda:0" if torch.cuda.is_available() else "cpu"

    print(f"Training on device: {device}")

    results = model.train(
        data=r"/content/yolo_dataset/data.yaml",
        epochs=50,
        imgsz=640,
        device=device,
        project="traffic_monitoring",
        name="yolo26_traffic",
        freeze=10,
        lr0=0.001,
        batch=16,
        verbose=True
    )

    print("Training complete. Results saved to 'traffic_monitoring/yolo26_traffic'")

    # # Export the model to preferred format (engine for TensorRT or just keep .pt)k
    # # Export to TensorRT
    # model.export(
    # format="engine",
    # imgsz=640,
    # half=True,
    # device=0
    # )

if __name__ == "__main__":
    train_model()
```

```python
import zipfile
import os

def zip_folder(folder_path, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, folder_path))

# Specify the folder to zip and the output zip file name
folder_to_zip = '/content/runs'
output_zip_file = '/content/runs.zip'

zip_folder(folder_to_zip, output_zip_file)
print(f"Folder '{folder_to_zip}' successfully zipped to '{output_zip_file}'")
```
